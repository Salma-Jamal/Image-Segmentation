```python
import torch
from PIL import Image
# import clip
import matplotlib.pyplot as plt
import numpy as np
import os
import numpy as np
import nibabel as nib
from sklearn.model_selection import train_test_split
import csv
from tqdm import tqdm
import random
import open_clip
import torch
from torch.utils.data import DataLoader, Dataset
import os
import random
import numpy as np
from tqdm import tqdm
import pandas as pd
```

```python
nii = nib.load('ACDC/database/training/patient001/patient001_frame12.nii.gz')
img = nii.get_fdata()

nii_gt = nib.load('ACDC/database/training/patient001/patient001_frame12_gt.nii.gz')
gt = nii_gt.get_fdata()
```

```python
import matplotlib.pyplot as plt

cols, rows = 2, 10  # Swapped cols and rows for better layout
fig, axs = plt.subplots(rows, cols, figsize=(4, 20))

for j in range(rows):
    axs[j, 0].imshow(img[:, :, j], cmap='gray')  # Image row
    axs[j, 0].axis('off')

    axs[j, 1].imshow(gt[:, :, j], cmap='gray')  # Ground truth row
    axs[j, 1].axis('off')

plt.tight_layout()
plt.show()
```
